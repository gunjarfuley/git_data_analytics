```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lrincome,urincome=remove_outlier(k1_scaled['spending_MinMaxScale'])
k1_scaled['spending_MinMaxScale']=np.where(k1_scaled['spending_MinMaxScale']>urincome,urincome,k1_scaled['spending_MinMaxScale'])
k1_scaled['spending_MinMaxScale']=np.where(k1_scaled['spending_MinMaxScale']<lrincome,lrincome,k1_scaled['spending_MinMaxScale'])
```

```python
lrincome,urincome=remove_outlier(k1_scaled['advance_payments_MinMaxScale'])
k1_scaled['advance_payments_MinMaxScale']=np.where(k1_scaled['advance_payments_MinMaxScale']>urincome,urincome,k1_scaled['advance_payments_MinMaxScale'])
k1_scaled['advance_payments_MinMaxScale']=np.where(k1_scaled['advance_payments_MinMaxScale']<lrincome,lrincome,k1_scaled['advance_payments_MinMaxScale'])
```

```python
lrincome,urincome=remove_outlier(k1_scaled['probability_of_full_payment_MinMaxScale'])
k1_scaled['probability_of_full_payment_MinMaxScale']=np.where(k1_scaled['probability_of_full_payment_MinMaxScale']>urincome,urincome,k1_scaled['probability_of_full_payment_MinMaxScale'])
k1_scaled['probability_of_full_payment_MinMaxScale']=np.where(k1_scaled['probability_of_full_payment_MinMaxScale']<lrincome,lrincome,k1_scaled['probability_of_full_payment_MinMaxScale'])
```

```python
lrincome,urincome=remove_outlier(k1_scaled['current_balance_MinMaxScale'])
k1_scaled['current_balance_MinMaxScale']=np.where(k1_scaled['current_balance_MinMaxScale']>urincome,urincome,k1_scaled['current_balance_MinMaxScale'])
k1_scaled['current_balance_MinMaxScale']=np.where(k1_scaled['current_balance_MinMaxScale']<lrincome,lrincome,k1_scaled['current_balance_MinMaxScale'])
```

```python
lrincome,urincome=remove_outlier(k1_scaled['credit_limit_MinMaxScale'])
k1_scaled['credit_limit_MinMaxScale']=np.where(k1_scaled['credit_limit_MinMaxScale']>urincome,urincome,k1_scaled['credit_limit_MinMaxScale'])
k1_scaled['credit_limit_MinMaxScale']=np.where(k1_scaled['credit_limit_MinMaxScale']<lrincome,lrincome,k1_scaled['credit_limit_MinMaxScale'])
```

## Apply K-Means clustering on scaled data

```python
from sklearn.cluster import KMeans 
```

```python
k_means = KMeans(n_clusters = 1)
k_means.fit(k1_scaled)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 2)
k_means.fit(k1_scaled)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 3)
k_means.fit(k1_scaled)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 4)
k_means.fit(k1_scaled)
k_means.inertia_
```

```python
wss =[]
```

```python
for i in range(1,11):
        KM = KMeans(n_clusters=i)
        KM.fit(k1_scaled)
        wss.append(KM.inertia_)
```

```python
wss
```

```python
plt.plot(range(1,11), wss)
plt.show()
```

```python
k_means_4 = KMeans(n_clusters = 4)
k_means_4.fit(k1_scaled)
labels_4 = k_means_4.labels_
```

```python
kmeans4_dataset=k1.copy()
```

```python
kmeans4_dataset["Cluster_kmeans"] = labels_4
kmeans4_dataset.head(5)
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
silhouette_score(k1_scaled,labels_4)
```

```python
from sklearn import metrics
```

```python
final = []
k_range = range(2, 11)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=2)
    km.fit(k1_scaled)
    final.append(metrics.silhouette_score(k1_scaled, km.labels_))
    
final
```

```python
plt.plot(k_range,final)
plt.show()
```
