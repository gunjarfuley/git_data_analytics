### Outlier Treatment

```python
plt.figure(figsize = (20,8))
b.boxplot(rot = 90);
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lr= Q1-(1.5 * IQR)
    ur= Q3+(1.5 * IQR)
    return lr, ur
```

```python
for column in b.columns:
    if (b[column].dtype != 'O') :
        b[column] = b[column].astype(int)
        lr,ur = remove_outlier(b[column])
        b[column]=np.where(b[column]>ur,ur,b[column])
        b[column]=np.where(b[column]<lr,lr,b[column])
```

### Laptop Logistic Regression

```python
#The split done is of 80: 20 ratio. 80% is Train & 20% is Test data
from sklearn.linear_model import LogisticRegression
X_train_LR, X_test_LR, y_train_LR, y_test_LR = train_test_split(X_sm, y_sm, test_size=0.2, random_state=42)
model_LR = LogisticRegression()
model_LR.fit(X_train_LR, y_train_LR)
```

```python
print('The model score for Logistic Regression Training set is',model_LR.score(X_train_LR,y_train_LR))
print('\n')
print('The model score for Logistic Regression Testing set is',model_LR.score(X_test_LR,y_test_LR))
```

```python
y_train_pred_LR = model_LR.predict(X_train_LR)
y_test_pred_LR = model_LR.predict(X_test_LR)
```
