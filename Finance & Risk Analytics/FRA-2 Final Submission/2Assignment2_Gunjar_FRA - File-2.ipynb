#### Importing the dataset

```python
stock_prices = pd.read_csv('Market+Risk+Dataset.csv')

#Glimpse of Data
stock_prices.head()
```

```python
stock_prices.info()
```

We can see that the datatype for Date is object.
We need to change format of 'Date' variable from object to Datetime.

```python
stock_prices['Date'] = [pd.to_datetime(d) for d in stock_prices['Date']]
```

```python
plt.figure(figsize=(10,5))
plt.scatter(stock_prices['Date'], stock_prices['SAIL'], edgecolors='r')
plt.xlabel("Date")
plt.ylabel("SAIL Stock Price")
plt.title("SAIL Stock Price Graph")
```
